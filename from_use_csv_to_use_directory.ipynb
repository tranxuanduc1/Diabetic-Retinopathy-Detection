```python


# Nếu không có sklearn, dùng manual stratified split như sau

# Đường dẫn
train_dir = '/home/duc/Documents/DoAn/messidor/preprocess'
output_dir = '/home/duc/Documents/DoAn/messidor/train_1'  # Thư mục output mới

# Đọc CSV
df = pd.read_csv("/home/duc/Documents/DoAn/messidor/messidor_data.csv")

# Manual stratified split
df = df.sort_values('diagnosis')



for level in range(5):
    dir_path = os.path.join(output_dir, str(level))
    os.makedirs(dir_path, exist_ok=True)

# Xử lý ảnh
for _, row in df.iterrows():
    image_name = row['id_code']
    level = row['diagnosis']
    
    # src_path = os.path.join(train_dir, f"{image_name}.jpeg")
    src_path = os.path.join(train_dir, f"{image_name}")
    
    if not os.path.exists(src_path):
        print(f"File not found: {src_path}")
        continue
    
    # Open PNG and save as JPG
    img = Image.open(src_path)
    file_name, file_extension = os.path.splitext(image_name)
    dest_dir = os.path.join(output_dir, str(level))
    dest_path = os.path.join(dest_dir, f"{file_name}.jpg")
    img.convert('RGB').save(dest_path, 'JPEG')
    
    print(f"Converted and moved {file_name}.jpeg to {dest_path}")

# Optional: Xóa ảnh gốc nếu cần, nhưng không khuyến khích
# shutil.rmtree(train_dir)  # Cẩn thận!
```
