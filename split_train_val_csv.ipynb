```python
import pandas as pd
import os
import shutil
import random

```

```python

train_dir = 'eyepascs_2015/train_split'
val_dir = 'eyepascs_2015/val_split'
csv_path = 'eyepascs_2015/trainLabels.csv'


os.makedirs(val_dir, exist_ok=True)


df = pd.read_csv(csv_path)

# Thêm phần mở rộng cho tên file
df['filename'] = df['image'].astype(str) + '.jpeg'

# Danh sách lưu lại dữ liệu mới
train_records = []
val_records = []
```

```python
df['level'] == sorted(df['level'].unique())[0]
```

```python
df[df['level'] == sorted(df['level'].unique())[1]]
```

```python

# Lặp qua từng nhãn (level)
for label in sorted(df['level'].unique()):
    # Lấy tất cả ảnh có nhãn hiện tại
    label_df = df[df['level'] == label]
    
    # Chọn ngẫu nhiên 200 ảnh (nếu đủ)
    sampled_df = label_df.sample(n=200, random_state=42)

    for _, row in sampled_df.iterrows():
        src_path = os.path.join(train_dir, row['filename'])
        dst_path = os.path.join(val_dir, row['filename'])
        
        # Di chuyển ảnh sang thư mục validation
        if os.path.exists(src_path):
            shutil.move(src_path, dst_path)
            val_records.append(row)
        else:
            print(f"Ảnh không tồn tại: {src_path}")

    # Các ảnh còn lại giữ lại cho training
    remaining_df = label_df.drop(sampled_df.index)
    train_records.extend(remaining_df.to_dict(orient='records'))

# Gộp lại thành DataFrame và lưu file CSV
train_df_new = pd.DataFrame(train_records)
val_df_new = pd.DataFrame(val_records)

train_df_new.to_csv('train_updated.csv', index=False, columns=['image', 'level'])
val_df_new.to_csv('validation_updated.csv', index=False, columns=['image', 'level'])

print("Đã hoàn tất chuyển ảnh và tạo file CSV.")
```
